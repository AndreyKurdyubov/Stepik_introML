```python
sns.lmplot(x='math_score', y='reading_score', data=students_performance)
```

добавим группирующую переменную:

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance)
```

без регрессионой прямой:

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg = False)
```

кастомизация:

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg = False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```

```python
data = students_performance.copy()
```

```python
sns.pairplot(data, hue='race/ethnicity')
```

```python
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs):
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    print(coef)
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    print(label)
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy=(0.2, 0.45), size=20, xycoords=ax.transAxes)

g = sns.PairGrid(data, hue='race/ethnicity', diag_sharey=False)
# g.map_upper(sns.scatterplot)
g.map_upper(corr)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
```
